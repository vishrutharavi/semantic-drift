```python
import os

news_dir = "/content/drive/MyDrive/news_crawl"

files = [
    os.path.join(news_dir, f)
    for f in os.listdir(news_dir)
    if f.endswith(".gz")
]

files = sorted(files)

print(files)
```

```python
import gzip
import re

def extract_news_crawl_contexts(word, files, max_samples=1500):

    pattern = re.compile(rf"\b{re.escape(word)}\b", re.IGNORECASE)

    contexts = []

    for file in files:

        print("Processing:", file)

        with gzip.open(file, "rt", encoding="utf8", errors="ignore") as f:

            for line in f:

                line = line.strip()

                if not line:
                    continue

                if len(line) > 400:
                    continue

                if pattern.search(line):

                    contexts.append(line)

                    if len(contexts) >= max_samples:
                        return contexts

    return contexts
```

```python
news_contexts = extract_news_crawl_contexts(
    "cloud",
    files,
    max_samples=1500
)

len(news_contexts)
```

```python
import re

def filter_old_cloud_contexts(contexts):

    bad_patterns = [
        r"\bcloud computing\b",
        r"\bthe cloud\b",
        r"\bcloud service\b",
        r"\bcloud platform\b",
        r"\bcloud storage\b",
        r"\bcloud provider\b",
        r"\bcloud infrastructure\b",
        r"\bcloud network\b",
        r"\bcloud system\b",
        r"\bcloud application\b",
        r"\bcloud computing\b",
        r"\bcloud-based\b",
        r"\bcloud provider\b"
    ]

    cleaned = []

    for s in contexts:

        bad = any(re.search(p, s, flags=re.IGNORECASE) for p in bad_patterns)

        if not bad:
            cleaned.append(s)

    return cleaned
```

```python
news_contexts = filter_old_cloud_contexts(news_contexts)

print(len(news_contexts))

for c in news_contexts[:10]:
    print("-", c)
```
